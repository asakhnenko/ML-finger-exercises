```python
import numpy as np
import matplotlib.pyplot as plt
```

## Plot slope fields and isocline

```python
def plotSlopeField(function, plot_isoclines=False):
    v = np.linspace(-5, 5, 30) 
    t = np.linspace(-5, 5, 30) 
    t_grid , v_grid = np.meshgrid(t, v)

    values = function(v_grid, t_grid)
    
    lng = np.sqrt(values**2 + 1)
    dt = 1 / lng
    dv = values / lng

    plt.xlabel('t(s)')
    plt.ylabel('V(V)')
    plt.quiver(t, v, dt, dv, label='Slope field', color='xkcd:azure')
    
    if plot_isoclines:
        isoclines = [-2, -1, 0, 1, 2]
        contours = plt.contour(t_grid, v_grid, values, isoclines)

        labels = ['isocline ' + str(iso) + ' v/s' for iso in isoclines]
        for i in range(len(labels)):
            contours.collections[i].set_label(labels[i])

        plt.legend(loc='upper right')
    plt.show()
```

$$\frac{dV}{dt} = 1 - V - t$$

```python
diff_eq_1 = lambda V, t : 1 - V - t
plotSlopeField(diff_eq_1, True)
```

$$\frac{dV}{dt} = sin(t) - \frac{1}{1.5}V$$

```python
diff_eq_2 = lambda V, t : np.sin(t) - V / 1.5 
plotSlopeField(diff_eq_2, True)
```

### Leaky integrate and fire neuron.

Derive the differential equation for the following equivalent circuit of a leaky integrate and fire neuron

<img src="circuit.png" style="height:100px">

According to Kirchhoff's current law: $$I_{ex} = I_R + I_C$$
Now we can substitute currents with known formulas. It is given that $I_{ex} = I_{max} sin(t)$. According to Ohm's law, current for resistor is $I_R = \frac{V}{R}$ and for capacitor is $I_C=C \frac{dV}{dt}$. From here is follows: $$I_{max} sin(t) = \frac{V}{R} + C \frac{dV}{dt}$$

$$\frac{dV}{dt} = \frac{1}{C}(I_{max} sin(t) - \frac{V}{R})$$

```python
diff_eq_3 = lambda V, t: (0 * np.sin(t) - V / 1.3) / 0.8
plotSlopeField(diff_eq_3)
```

```python
diff_eq_4 = lambda V, t: (1 * np.sin(t) - V / 1.3) / 0.8
plotSlopeField(diff_eq_4)
```

```python
diff_eq_5 = lambda V, t: (0 * np.sin(t) - V / 1.3) / 0.8 + 2
plotSlopeField(diff_eq_5)
```

```python
diff_eq_6 = lambda V, t: (1 * np.sin(t) - V / 1.3) / 0.8 + 2
plotSlopeField(diff_eq_6)
```
