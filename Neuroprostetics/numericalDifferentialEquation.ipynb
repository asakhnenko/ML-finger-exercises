### Numerical Solvers' Implementations for ODEs 

```python
import matplotlib.pyplot as plt

def plot(func, example, steps):
    plt.figure(figsize=(10,4))
    
    plt.xlim(-4.5, 5.5) 
    plt.ylim(-4.5, 6)
    plt.xticks(np.arange(-5, 5, step=1))
    plt.yticks(np.arange(-4, 6, step=2))

    for step in steps:
        example["step"] = step
        y, t = func(**example)
        plt.plot(t, y, label='$\Delta$ t = ' + str(example['step']) + ' s')

    plt.legend(loc='upper right')
    plt.grid(True)
    plt.xlabel('t (s)', fontsize = 10)
    plt.ylabel('V (V)', fontsize = 10)

    plt.show()
```

#### Forward (Explicit) Euler

Update rule: $$y_{n+1} = y_{n} + h f(t_n, y_n)$$

```python
import numpy as np
import math

def calculateForwardEuler(func, y_start, step, time_start, time_end):
    # Initialization
    t_max = int((time_end - time_start) / step)
    Y = np.array([y_start] * (t_max + 1))
    T = np.linspace(0, t_max, t_max + 1).astype('int')

    # Iteration
    for t in T[1:]:
        Y[t] = Y[t-1] + step * func((t-1) * step + time_start, Y[t-1])
    return Y, T * step + time_start
```

```python
steps = [1, 0.5, 0.1, 0.012]
example = {
    "func"       : lambda t, V: 1 - V - t,
    "y_start"    : -4.0,
    "time_start" : -4.5,
    "time_end"   : 5
}

plot(calculateForwardEuler, example, steps)
```

#### Heun method
Update rule:
$$\hat{y}_{n+1} = y_{n} + h f(t_n, y_n)$$
$$y_{n+1} = y_{n} + \frac{h}{2} (f(t_n, y_n) + f(t_{n+1}, \hat{y}_{n+1}))$$

```python
def calculateHeun(func, y_start, step, time_start, time_end):
    # Initialization
    t_max = int((time_end - time_start) / step)
    Y = np.array([y_start] * (t_max + 1))
    T = np.linspace(0, t_max, t_max + 1).astype('int')

    # Iteration
    for t in T[1:]:
        y_pred = Y[t-1] + step * func((t-1) * step + time_start, Y[t-1])
        f_diff = func((t-1) * step + time_start, Y[t-1]) + func(t * step + time_start, y_pred)
        Y[t] = Y[t-1] + step / 2 * f_diff
    return Y, T * step + time_start
```

```python
plot(calculateHeun, example, steps)
```

#### Exponential Euler

Given a differential equation $\frac{dV}{dt} = A(t)V(t) + B(V, t)$

An update rule is $V_{n+1} = V_{n} e^{A(t_n)h} + \frac{B(V_n, t_n)}{A(t_n)} (e^{A(t_n)h} - 1)$

```python
def calculateExponentialEuler(func, y_start, step, time_start, time_end):
    # Initialization
    t_max = int((time_end - time_start) / step)
    Y = np.array([y_start] * (t_max + 1))
    T = np.linspace(0, t_max, t_max + 1).astype('int')

    # --- for a specific example ----
    A = -1
    B = 1 - (T * step + time_start)
    # -------------------------------
    
    # Iteration
    for t in T[1:]:
        Y[t] = Y[t-1] * np.exp(A * step) + B[t-1] / A * (np.exp(A * step) - 1)
    return Y, T * step + time_start
```

```python
plot(calculateExponentialEuler, example, steps)
```

### Leaky Integrate and Fire Neuron

\begin{equation}
V_{n+1}=\begin{cases}
V_n + \frac{\Delta t}{C_m}(-g_{leak}(V_n-V_{rest})+I_{input}(t_n))& V_n < V_{thr}\\
V_{spike}&V_{thr} \leq V_n < V_{spike}\\
V_{rest}&V_{spike}  \leq V_{n}\end{cases}
\end{equation}

```python
def leakyIntegrateAndFire(I_input, C_m, g_leak, V_rest, V_thr, V_spike, step, duration):
    # Initialization
    t_max = int(duration / step)
    V = np.array([V_rest] * t_max)
    T = np.linspace(0, t_max - 1, t_max).astype('int')

    # Iteration
    for t in T[1:]:
        if V[t-1] < V_thr:
            V[t] = V[t-1] + (step / C_m) * (-g_leak * (V[t-1] - V_rest) + I_input[t-1])
        elif V_thr <= V[t-1] < V_spike:
            V[t] = V_spike
        elif V_spike <= V[t-1]:
            V[t] = V_rest

    return V, T * step
```

```python
example = {
    "C_m" : 0.000001,
    "g_leak" : 0.0001,
    "V_rest" : -0.06,
    "V_thr" : -0.02,
    "V_spike" : 0.02,
    "duration" : 0.05,
    "step" : 0.000025,
    "I_input" : [0.00002]*int(0.05 / 0.000025)
}
example["I_input"] = [0.00001] * int(0.05 / 0.000025)
V, t = leakyIntegrateAndFire(**example)

ax = plt.axes()
ax.tick_params(axis='both', which='major', pad=10, labelsize=10,
               direction='in') 
plt.xlim(-0.0025, 0.0525) 
plt.ylim(-0.0625, 0.0225)

plt.plot(t, V, linewidth=1.3)

plt.xlabel('t (s)', fontsize=10)
plt.ylabel('V (V)', fontsize=10)

plt.tight_layout()
```
